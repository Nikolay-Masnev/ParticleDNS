```python
import numpy as np 
import matplotlib.pyplot as plt
  
# Text file data converted to integer data type 
y = np.loadtxt("w_autocorr.data", dtype=float)[:100]
x = np.arange(start=0, stop=100, step=1)

plt.grid()
plt.xlabel(r'$\tau/10$')
plt.ylabel(r'$R(t)$')
plt.title(r'Autocorrelation at r = $3\pm 0.1$ cm')

plt.plot(x, y)
plt.show()
```

```python
y = np.loadtxt("vel_pdf.data", dtype=float)[:10]
x = np.arange(start=0, stop=10, step=1)/100

plt.grid()
plt.xlabel(r'$w_r$')
plt.ylabel(r'$P(w_r)$')
plt.title('Radial velocity PDF at r = 3cm')

plt.plot(x, y)
plt.show()
```

```python
y = np.loadtxt("phi_corr.data", dtype=float)
x = np.arange(start=0, stop=1000, step=1)

y = y[1:]
x = x[1:]

plt.grid()
plt.xlabel(r'$\tau/10$')
plt.ylabel(r'$\langle \phi(t)  - \phi(t + \tau)\rangle$')
plt.title(r'r = $3\pm 0.1$ cm')

plt.plot(x, y)
plt.show()
```

```python
y = np.loadtxt("phi_traj.data", dtype=float)
x = np.arange(start=0, stop=y.size, step=1)

plt.grid()
plt.xlabel(r'$\tau/10$')
plt.ylabel(r'$\phi(t)$')

plt.plot(x, y)
plt.show()
```
